```python
import cv2

img = cv2.imread(r"C:\Users\igorm\Documents\programacao\britago-rasp\frames_debug\frame_20260721_171402_056.jpg")  # troca pelo nome do frame
clone = img.copy()

def click(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        print(f"x={x}, y={y}")

cv2.namedWindow("frame")
cv2.setMouseCallback("frame", click)

while True:
    cv2.imshow("frame", clone)
    if cv2.waitKey(1) == 27:  # ESC pra sair
        break
cv2.destroyAllWindows()
```
